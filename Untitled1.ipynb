```python
import numpy as np
import pandas as pd
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt
import nltk
```

```python
data = pd.read_csv('JEOPARDY_CSV.csv')
data.head()
```

```python
data.drop(data[data[' Value'] == 'None'].index,inplace=True)
```

```python
data['ValueNum'] = data[' Value'].apply(
    lambda value: int(value.replace('$', '').replace(',','')))
```

```python
def binning(value):
    if value < 1000:
        return np.round(value, -2)
    elif value < 10000:
        return np.round(value, -3)
    else:
        return np.round(value, -4)

data['ValueBins'] = data['ValueNum'].apply(binning)
```

```python
data
```

```python
data.reset_index(inplace=True)
```

```python
data.drop('index',axis=1,inplace=True)
```

```python
data
```

```python
dataset = data[[' Question','ValueBins']]
dataset
```

```python
comment_words = '' 
stopwords = set(STOPWORDS) 

for val in data[' Question']: 
      
    val = str(val) 
    tokens = val.split() 

    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import re
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
corpus=[]

for i in range(len(dataset)):
    review = re.sub('[^a-zA-Z0-9]', ' ', dataset[' Question'][i])
    review = review.lower()
    review = review.split()
    review = [stemmer.stem(word) for word in review if not word in set(stopwords.words('english'))]  
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[1]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(max_features=5000)
X = tf.fit_transform(corpus).toarray()
```

```python
X
```

```python
y = dataset['ValueBins'].values
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect = MultinomialNB()
spam_detect.fit(X_train,y_train)
```

```python
y_pred = spam_detect.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)

# Accuracy Score
score = accuracy_score(y_test,y_pred)
```

```python
score
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_rf_pred = rf.predict(X_test)
y_rf_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm_rf = confusion_matrix(y_test,y_rf_pred)

# Accuracy Score
score_rf = accuracy_score(y_test,y_rf_pred)
```
